```python
from skimage import data,filters,util
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = data.camera()
img_filtered = filters.sobel_h(img)
target = img_filtered>0.07
plt.subplot(121), plt.imshow(img_filtered)
plt.subplot(122), plt.imshow(target)
```

```python
# scipy.signal.convolve2d

?filters.sobel_h

kernel_h = [[1,2,1],[0,0,0],[-1,-2,-1]]

from scipy.signal import convolve2d
test = convolve2d(img,kernel_h,mode='same')
plt.imshow(test>65)
```

```python
# scipy.ndi.convolve

from scipy import ndimage
test2 = ndimage.convolve(img, kernel_h)
plt.imshow(test2>65)
#tried with difference thresholding
```
